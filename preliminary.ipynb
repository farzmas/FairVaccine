```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/COVID-19_Daily_Vaccinations_-_Chicago_Residents.csv')
b = data['Total Doses - Daily - Black Non-Latinx'].values
w = data['Total Doses - Daily - White Non-Latinx'].values
l = data['Total Doses - Daily - Latinx'].values
a = data['Total Doses - Daily - Asian Non-Latinx'].values
plt.plot(b, label ='Black')
plt.plot(w, label = 'White')
plt.plot(l, label = 'Latinx')
plt.plot(a, label = 'Asian')
plt.legend()
plt.title('Chicago daily vaccination')
plt.show()
```
